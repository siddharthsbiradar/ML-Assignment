```python
mu,sigma=5,1
x = np.random.normal(mu,sigma,10)
np.mean(x)
```

```python
def get_liklihood(z):
    ans= 1/(np.sqrt(2*np.pi*sigma**2)) * np.exp(-0.5*((x-z)/sigma)**2)
    return ans
```

```python
def gaussian(x, mean, variance):
    const = 1/(((2*math.pi)**0.5)*sigma)
    y = math.e**((-0.5*(sigma**2))*((x - mean)**2))
    return y*const

def likelihood(x, mean):
    p = 1
    for i in x:
        p *= gaussian(i, mean, 1)
    return p
```

```python
means = [0]*11
likelihoods = [0]*11
for i in range(0, 11):
    likelihoods[i] = likelihood(x, i)
    means[i] = i
```

```python
sns.lineplot(means, likelihoods)
plt.xlabel("Mean")
plt.ylabel("Likelihood")
plt.title("Likelihood curve")
plt.show()
```
